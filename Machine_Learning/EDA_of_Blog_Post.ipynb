**Importing Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
```

##Loading Zomato_Restaurant Data Set

```python
zds1=pd.read_csv('/content/zomato.csv')
print(zds1)
```

**Taking a sample 50 rows**

```python
zds = zds1.sample(50)
zds1.head()
```

**Handling Missing Values of Zomato DataSet.**

```python
rate_mode = zds['rate (out of 5)'].mean()
zds['rate (out of 5)'] = zds['rate (out of 5)'].fillna(rate_mode)
zds = zds.dropna()
print(zds.isnull().sum())
```

##Data visualization
* Distribution of ratings

```python
plt.figure(figsize=(10, 6))
sbn.histplot(zds['rate (out of 5)'],bins=30, kde=True)
plt.title('Distribution of Ratings')
plt.show()
```

##Pie Chart

```python
location_distribution = zds['area'].value_counts()
plt.figure(figsize=(12,8))
plt.pie(location_distribution, labels=location_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Restaurants by Location\n')
plt.axis('equal')
plt.show()
```

##Number of Restaurants located by City
* Plotting the information about the count of restaurants locations using Count plot

```python
plt.figure(figsize=(12, 8))
sbn.countplot(y='area', data=zds, order=zds['area'].value_counts().index)
plt.title('Count of Restaurants by City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

**Let's calculate the average rating by City and plot it using Bar_Plot**

```python
plt.figure(figsize=(12, 8))
avg_rating_city = zds.groupby('area')['rate (out of 5)'].mean().sort_values()
sbn.barplot(x=avg_rating_city, y=avg_rating_city.index)
plt.title('Average Rating by City')
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.show()
```

##Correlation Heatmap
* We create a heatmap to visualize correlations between numeric features.


```python
numeric_cols = zds.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()
plt.figure(figsize=(10, 8))
sbn.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
