##Handling Missing Values

```python
cds['mileage'] = cds['mileage'].fillna(cds['mileage'].mean())
print(cds.isnull().sum())
```

##Data Visualization
* Distribution of Price
* plotting the distribution of the Price column.

```python
plt.figure(figsize=(10, 6))
sbn.histplot(cds['price'], bins=30, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

##Scatter Plot of Mileage vs. Price

* Creating a scatter plot to show the relationship between mileage and price.

```python
plt.figure(figsize=(10, 6))
sbn.scatterplot(x='mileage', y='price', data=cds)
plt.title('Mileage vs. Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

##Box Plot of Price by Brand
We'll create a box plot to show the distribution of price by brand.

```python
plt.figure(figsize=(10, 6))
sbn.boxplot(x='brand', y='price', data=cds)
plt.title('Box Plot of Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

##Pair Plot of Key Features
Creating a pair plot to visualize the relationships between key features.

```python
sbn.pairplot(cds, vars=['price', 'mileage', 'year'], hue='brand')
plt.suptitle('Pair Plot of Key Features', y=1.02)
plt.show()
```

##Correlation Heatmap
Creating a heatmap to show the correlation between different features.

```python
numeric_cols = cds.select_dtypes(include=[np.number])
# Calculation of the correlation matrix
corr_matrix = numeric_cols.corr()
# Plotting the heatmap
plt.figure(figsize=(10, 8))
sbn.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

##Advanced Visualizations
* Distribution of Car Brands
* Plotting the distribution of Car brands.

```python
plt.figure(figsize=(12, 8))
sbn.countplot(y='brand', data=cds, order=cds['brand'].value_counts().index)
plt.title('Distribution of Car Brands')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()
```

##Pie Chart of Car Brands

A pie chart is a great way to visualize the proportion of different categories.

```python

brand_distribution = cds['brand'].value_counts()
# Create a pie chart
plt.figure(figsize=(12, 8))
plt.pie(brand_distribution, labels=brand_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Car Brands')
plt.axis('equal')
plt.show()
```
