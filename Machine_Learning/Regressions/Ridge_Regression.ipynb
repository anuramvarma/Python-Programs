```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import Ridge

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
stups = pd.read_csv('/content/50_Startups.csv')
stups.sample(20)
```

**Checking the features of Data Set**

```python
stups.columns
```


**checking Information that present in Data Set**

```python
stups.info()
```

**Checking the null values.**

If we found null values we have to adjust the data before to train the model

 **Here drop the states column**

```python
stups.isnull().sum()
data = stups.drop(['State'],axis=1)
data.head()
```

```python
p = data.drop(['Profit'],axis=1).values
q = data['Profit'].values
```

```python
p_train , p_test , q_train , q_test = train_test_split(p,q,test_size=0.25,random_state=42)
p_train.shape , q_test.shape
```

```python
reg = Ridge(alpha=0.1)
reg.fit(p_train,q_train)
```

```python
q_pred1 = reg.predict(p_train)
print(f'Performance of the model on training data :\n')
print(f'MAE = {mean_absolute_error(q_train, q_pred1)}')
print(f'MSE = {mean_squared_error(q_train, q_pred1)}')
print(f'RMSE = {np.sqrt(mean_squared_error(q_train, q_pred1))}')
print(f'R_2 = {r2_score(q_train, q_pred1)}')
```

```python
q_pred2 = reg.predict(p_test)
print(f'Performance of the model on test data :\n')
print(f'MAE = {mean_absolute_error(q_test, q_pred2)}')
print(f'MSE = {mean_squared_error(q_test, q_pred2)}')
print(f'RMSE = {np.sqrt(mean_squared_error(q_test, q_pred2))}')
print(f'R_2 = {r2_score(q_test, q_pred2)}')
```

```python
reg.score(p_test,q_test)
```
