###Linear_Regression


Importing the packages that are required to perfotm or work with regressions.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import sklearn.linear_model as sklm
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

Loading the Data Set i.e 50 Startups

```python
stups = pd.read_csv('/content/50_Startups.csv')
stups.sample(20)
```

Checking the features of Data Set

```python
stups.columns
```

**checking Information that present in Data Set**

```python
stups.info()
```

**Checking the null values.**

* If we found null values we have to adjust the data before to train the model

**Here drop the states column**

```python
stups.isnull().sum()
data = stups.drop(['State'],axis=1)
data.head()
```

```python
p = data.drop(['Profit'],axis=1).values
q = data['Profit'].values
reg = sklm.LinearRegression()
reg.fit(p,q)
```

##Linear Regression

```python
q_pred = reg.predict(p)
print(f'MAE = {mean_absolute_error(q, q_pred)}')
print(f'MSE = {mean_squared_error(q, q_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(q, q_pred))}')
print(f'R_2 = {r2_score(q, q_pred)}')
```

```python
reg.score(p,q)
```
