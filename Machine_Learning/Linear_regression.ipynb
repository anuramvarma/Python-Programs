First importing the necessary packages
In this model I am doing regression on 50_startups using Lasso

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model as lm

from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

Reading the csv file

Checking the data whether all rows have numeric values or not
Because we want all numerical values to train the model

```python
su = pd.read_csv('/content/50_Startups.csv')
su.head()
```

Checking the null values.

(If we found null values we have to adjust
the data before to train the model)

Here drop the states column

```python
su.isnull().sum()
data = su.drop(['State'],axis=1)
data.head()
```

```python
x = data.drop(['Profit'],axis=1).values
y = data['Profit'].values
```

```python
reg = lm.LinearRegression()
reg.fit(x,y)
```

```python
y_pred = reg.predict(x)

print(f'MAE = {mean_absolute_error(y, y_pred)}')
print(f'MSE = {mean_squared_error(y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y, y_pred))}')
print(f'R_2 = {r2_score(y, y_pred)}')
```

```python
reg.score(x,y)
```
